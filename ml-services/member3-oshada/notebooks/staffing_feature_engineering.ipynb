# Member 3: Feature Engineering & Model Comparison

This notebook explores different feature combinations and models for optimizing employee staffing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import cross_val_score

data_path = '../data/employee_demand_dataset.csv'
df = pd.read_csv(data_path)
print(f"Loaded {len(df)} records.")
```

## 1. Feature Importance Analysis

```python
# Quick Random Forest to check importance
from sklearn.preprocessing import LabelEncoder

df_clean = df.copy()
le = LabelEncoder()
df_clean['weather_enc'] = le.fit_transform(df_clean['weather'])

features = ['month', 'day_of_week', 'is_weekend', 'is_holiday', 'temperature', 'predicted_fuel_demand', 'weather_enc']
X = df_clean[features]
y = df_clean['employee_count']

rf = RandomForestRegressor(n_estimators=100)
rf.fit(X, y)

importance = pd.DataFrame({'feature': features, 'importance': rf.feature_importances_}).sort_values('importance', ascending=False)
sns.barplot(data=importance, x='importance', y='feature')
plt.title("Initial Feature Importance")
plt.show()
```
